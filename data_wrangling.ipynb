```python
colname_tuple = list(df.index[1])
colname_tuple[4] = 'Sunrise'
colname_tuple[5] = 'Sunculm'
colname_tuple[6] = 'Sunset'
colname_tuple[-3] = 'Moonrise'
colname_tuple[-2] = 'Moonculm'
colname_tuple[-1] = 'Moonset'
```

```python
data_list = []
for file in glob.glob('Escondido 2021 sunrise and sunset data/*.csv'):
    df = pd.read_csv(file)
    
    for row in list(np.asarray(df.index[2:-3].map(lambda x: list(x)))):
        data_list.append(row)
```

```python
len(data_list)
```

```python
data = pd.DataFrame(data_list, columns = ['Year', 'Month', 'Day', 'Age of Moon', 'Sunrise', 'Sunculm', 'Sunset', 'Constellation',
                                   'Moonrise', 'Moonculm', 'Moonset'])
```

```python
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Day']])
```

```python
data = data.sort_values('Date')
```

```python
data = data.drop('Date', axis = 1)
```

```python
data.index = range(data.shape[0])
```
